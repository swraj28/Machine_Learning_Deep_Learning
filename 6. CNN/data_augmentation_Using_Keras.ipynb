## Data Augmentation
- Deep Learning Models are "Data Hungry"
- Small Datasets lead to overfitting most of the times
- So lets generate more data by augmentating the training samples

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img_path=r"C:\Users\SWRAJ KANT SHARMA\Documents\Datasets\Dataset\Images\images\cats\cat.1.jpg"
```

```python
img = image.load_img(img_path,target_size=(200,200))
img = image.img_to_array(img)
print(type(img))
```

```python
x = img/255.0
print(x.shape)
plt.imshow(x)
plt.axis('off')
plt.show()
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
            rotation_range = 30,
            width_shift_range = 0.2,
            height_shift_range = 0.2,
            shear_range = 0.2,
            zoom_range = 0.2,
            horizontal_flip = 0.2,
            fill_mode = "nearest"
)
```

```python
input_batch = x.reshape((1,*x.shape))
print(input_batch.shape)
```

```python
i = 0
for output_batch in datagen.flow(input_batch,batch_size=1):
    plt.figure()
    imgplot = plt.imshow(image.img_to_array(output_batch[0]))
    i+= 1
    
    plt.axis("off")
    plt.show()
    
    if i==10:
        break
    
    
```
