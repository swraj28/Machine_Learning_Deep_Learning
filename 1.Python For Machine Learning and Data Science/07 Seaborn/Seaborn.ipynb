```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# plot the distribution of the data ---> like a histogram
sns.distplot(iris['petal_length'], bins= 20, kde=False)
plt.show()
```

```python
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x= iris['petal_length'], y = iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x ='petal_length', y ='petal_width' , data= iris, hue='species')
plt.show()
```

```python
sns.jointplot(x= 'petal_length', y = 'petal_width', data= iris, kind='reg')
plt.show()
```

```python
sns.jointplot(x= 'sepal_length', y = 'sepal_width', data= iris, kind='reg')
plt.show()
```

```python
#Now lets dig deeper and find the children among the passengers. The passenger whose age is 
# less than 16 will be considered as child
```

```python
def child_passenger(passenger):
    age,sex=passenger
    if age<16:
        return 'child'
    else:
        return sex
```

```python
titanic['person']=titanic[['age','sex']].apply(child_passenger,axis=1)
```

```python
titanic.head()
```

```python
### Plotting a graph to check the ratio of male,female and children in each
# category of class

sns.countplot(x="pclass",data=titanic,hue="person")
plt.show()
```
