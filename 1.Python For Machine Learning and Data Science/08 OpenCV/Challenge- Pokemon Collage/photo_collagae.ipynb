```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1=plt.imread('./img1.jpg')
img2=plt.imread('./img2.jpg')
img3=plt.imread('./img3.jpg')
img4=plt.imread('./img4.jpg')
img5=plt.imread('./img5.jpeg')
```

```python
plt.axis("off")
plt.imshow(img1)
plt.show()
```

```python
print(img1.shape)
print(img2.shape)
print(img3.shape)
print(img4.shape)
print(img5.shape)
```

```python
img1=cv2.resize(img1,(200,200))
print(img1.shape)
plt.axis("off")
plt.imshow(img1)
plt.show()
```

```python
img2=cv2.resize(img2,(200,200))
img3=cv2.resize(img3,(200,200))
img4=cv2.resize(img4,(200,200))
img5=cv2.resize(img1,(100,100))
```

```python
print(img1.shape)
print(img2.shape)
print(img3.shape)
print(img4.shape)
print(img5.shape)
```
