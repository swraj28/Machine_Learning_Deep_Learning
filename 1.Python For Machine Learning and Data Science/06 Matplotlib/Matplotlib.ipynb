## Line Plots

## Scatter Plot
 - plt.scatter()
 - scatters data points across x and y axis.

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y1, color ='red', label="iPhone", marker="*")
plt.scatter(x, y2, color='green', label="Android")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Smartphones over time")
plt.legend()
plt.show()
```

## Bar Graphs
 - plt.bar()
 - plot a bar graph between heights and labels you provides. 

```python
x = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]
```

```python
plt.bar(x, car1, width=0.3, tick_label= [2017, 2018, 2019, 2020], label="Car 1")
plt.bar(x+0.3, car2, width=0.3, label="Car 2")
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.xlim(-1, 5) #padding on the x axis
# plt.ylim()
plt.show()
```

## Pie Chart
- plt.pie()
- plot the pie chart with the provides labels and values

```python
subjects = ["Physics", 'Chemistry', 'Maths', "English"]
marks = np.array([78, 45, 90, 65])
```

```python
#autopct-> percentage of marks
plt.pie(x= marks, labels = subjects, explode = [0, 0.1, 0, 0], shadow= True, autopct = "%0.0f")
plt.show()
```

## Histograms
 - plt.hist()
 - creating dummy data using normal distribution 

```python
#mean=0 and standard deviation->1
# marks will vary  from -1 to +1
X_snd = np.random.randn(500)
```

```python
maths_mu = 70 # mean
maths_sig = 8 #std.deviation
maths_marks = np.round(X_snd*maths_sig + maths_mu)

phy_mu = 50
phy_sig = 10
phy_marks = np.round( X_snd*phy_sig + phy_mu )
```

```python
#historgram require only one parameter
plt.hist(maths_marks)
plt.hist(phy_marks, alpha = 0.7)  # alpha -> opacity
plt.show()
```

## Subplots
- create multiple plots in just a single figure.

```python
x = np.arange(0,10, 0.1)
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
plt.subplot(2,3,1)
plt.plot(x, y1)

plt.subplot(2,3,2)
plt.plot(x, y2)

plt.subplot(2,3, 3)
plt.plot(x, y3)

plt.subplot(2,3, 4)
plt.plot(x, y4)

plt.subplot(2,3, 5)
plt.plot(x, y3)

plt.subplot(2,3, 6)
plt.plot(x, y4)

plt.show()
```

## Challenge - Movie Data Visualization
Given movie Dataset, your task is to make a visualization that - 
 - Plots the length of Movie Title Names on X-Axis,
 - Frequency counts on Y-axis i.e Number of movies having x characters in their title


```python
movie = pd.read_csv("./movie_meta.csv")
```

```python
movie.head()
```

```python
movie.columns
```

```python
def preprocess(ele):
    return len(ele.strip()) # Returns a copy of element by removing trailing and leading space 
```

```python
movie['movie_title_len'] =  movie['movie_title'].apply(preprocess)
movie.shape
```

```python
movie.columns
```

```python
plt.hist(movie['movie_title_len'], bins = 50)
plt.xticks(range(0, 100, 5))
plt.xlabel("Movie title len")
plt.ylabel("Number of movies (Freq)")
plt.show()
```

### Scatter Way

```python
movie['movie_title_len']
```

```python
freq_title = {}

for l in movie['movie_title_len']:
    if freq_title.get(l) is None:
        freq_title[l] = 1
    else:
        freq_title[l] += 1
```

```python
x = freq_title.keys()
y = freq_title.values()
```

```python
plt.scatter(x, y)
plt.xticks(range(0, 60, 5))
plt.xlabel("Movie title len")
plt.ylabel("Number of movies (Freq)")
plt.show()
```
