```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dfx=pd.read_csv(r'C:\Users\SWRAJ KANT SHARMA\Documents\Datasets\KNN\Diabetis\Diabetes_XTrain.csv')
dfy=pd.read_csv(r"C:\Users\SWRAJ KANT SHARMA\Documents\Datasets\KNN\Diabetis\Diabetes_YTrain.csv")
```

```python
dfx.head(n=10)
```

```python
dfx=dfx.drop(['SkinThickness'],axis=1)
dfx=dfx.drop(['Pregnancies'],axis=1)
```

```python
dfy.head(n=10)
```

```python
X_train=dfx.values
print(X_train.shape)
print(X_train)
```

```python
Y_train=dfy.values
print(Y_train)
```

```python
### Normalization 
x_mean=X_train.mean(axis=0);
x_std=X_train.std(axis=0)
```

```python
X_train=(X_train-x_mean)/x_std
print(X_train.shape)
print(X_train)
```

### Implementation

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=25)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
df=pd.read_csv(r"C:\Users\SWRAJ KANT SHARMA\Documents\Datasets\KNN\Diabetis\Diabetes_Xtest.csv")
df=df.drop(['SkinThickness'],axis=1)
X_test=df.values
```

```python
X_test=(X_test-x_mean)/x_std
```

```python
Y_pred=model.predict(X_test)
```

```python
print(Y_pred)
```

```python
Y_pred.reshape((-1,1))
```

```python
df=pd.DataFrame(data=Y_pred,columns=["outcome"])
```
