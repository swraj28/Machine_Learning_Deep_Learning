```python
#Making numpy array X and Y
#X contains the datapoints(x and y cordinates) and Y contains the class of the datapoints.
X = dfx.values
Y = dfy.values


# The zeroth column contains serial no therefore we will discard it

X = X[:,1:]
Y = Y[:,1:].reshape((-1,))

#print(X)

print(X.shape)
print(Y.shape)
print(X)
print(Y)
```

```python
#KNN Algorithm:-

def dist(x1,x2): # Euclidian Distance
    return np.sqrt(sum((x1-x2)**2))

# Test Time (No Training Is Involved)
def knn(X,Y,queryPoint,k=5):
    
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(queryPoint,X[i])
        #vals contains 2 parametrs.
        # a.) distance of points from dataset from querypoint
        # b.) label of each point dataset
        vals.append((d,Y[i]))
        
    
    vals = sorted(vals)
    
    # Nearest/First K points
    vals = vals[:k]
    
    vals = np.array(vals)
    
#     print(vals)
    
    #To count the frqurncy of the two classes (Labels)
    new_vals = np.unique(vals[:,1],return_counts=True)
#     print(new_vals)
    
    #Index Of The Maximum Frequency
    index = new_vals[1].argmax()
#     print(index)

    pred = new_vals[0][index]
    
    return pred
```
