```python
import os, shutil
folders = os.listdir("Images")
```

```python
print(folders)
```

```python
image_data=[]
labels=[]

label_dict={
        "cats":0,
        "dogs":1,
        "horses":2,
        "humans":3
}
```

```python
from keras.preprocessing import image
for ix in folders:
  path=os.path.join("Images",ix)
  # print(path)
  # Extract all the images in that path
  for im in os.listdir(path):
    img=image.load_img(os.path.join(path,im),target_size=((224,224)))
    img_array=image.img_to_array(img)
    image_data.append(img_array)
    labels.append(label_dict[ix])
```

```python
print(len(image_data),len(labels))
```

```python
# Shuffle the data
import random
combined=list(zip(image_data,labels))
random.shuffle(combined)

image_data[:],labels[:]=zip(*combined)
```

```python
print(labels[:5])
```

```python
import numpy as np
X_train=np.array(image_data)
Y_train=np.array(labels)

print(X_train.shape,Y_train.shape)
```

```python
# from keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense,Activation,Flatten,Input,Dropout,GlobalAveragePooling2D
from tensorflow.keras.models import Model
from keras.utils.np_utils import to_categorical

import matplotlib.pyplot as plt
```

```python
# One hot label
Y_train=to_categorical(Y_train)
print(X_train.shape,Y_train.shape)
```

```python
# Create the Resnet50 model
# We are using only convolutional base not the clasifier base
model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
model.summary()
```

```python
# We are going to take global average pooling from each output layer which is going to
# take average from each layer and converts each channel to a single scalar.
# conv5_block3_out (Activation)   (None, 7, 7, 2048) ->converts to a single vector of 2048 dims
# This is done to reduce the parameters

av1 = GlobalAveragePooling2D()(model.output) #The input to this layer will be the output of the convolutional Base Of ResNet50

# MLP Classifier
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(4,activation='softmax')(d1)

# We are using the functional api of keras (Connecting The conv.Base to the Classifier Part)
model_new = Model(inputs=model.input, outputs= fc2)
model_new.summary()
```

```python
adam = Adam(lr=0.00003)
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
# Iterate over all the layers in the model
for ix in range(len(model_new.layers)):
    print(ix, model_new.layers[ix])
```

```python
# Every layer has a trainable property.we are going to freeze the model from
# layer 1 to 169 (i.e,not changing the weights)
# from 170 to 174 we are going to back propagate and fine tune the weights
for ix in range(169):
    model_new.layers[ix].trainable = False
    
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
model_new.summary()
```

```python
hist = model_new.fit(X_train,Y_train,
                    shuffle = True,
                    batch_size = 16,
                    epochs = 5,
                    validation_split=0.20
                    )
```

```python
plt.figure(figsize=(10,10))
plt.style.use("seaborn")
plt.plot(hist.history['accuracy'],'b',label="accuracy")
plt.plot(hist.history['val_accuracy'],'g',label="val_accuracy")
plt.plot(hist.history['loss'],'black',label="loss")
plt.plot(hist.history['val_loss'],'red',label="val_loss")
plt.legend()
plt.show()
```
