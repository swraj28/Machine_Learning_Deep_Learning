```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
plt.style.use("seaborn")
```

### KNN

```python
X,Y=make_blobs(n_samples=1000,centers=2,n_features=2)
```

```python
plt.scatter(X[:,0][Y==0],X[:,1][Y==0],label='Class 0',cmap=plt.cm.Accent)
plt.scatter(X[:,0][Y==1],X[:,1][Y==1],label='Class 1',cmap=plt.cm.Accent)
plt.legend()
plt.show()
```

```python
print(X.shape)
print(Y.shape)
```

```python
query=np.array([-3,2])
```

```python
plt.scatter(X[:,0][Y==0],X[:,1][Y==0],label='Class 0',cmap=plt.cm.Accent)
plt.scatter(X[:,0][Y==1],X[:,1][Y==1],label='Class 1',cmap=plt.cm.Accent)
plt.legend()
plt.scatter(query[0],query[1],color="red",cmap=plt.cm.Accent)
plt.show()
```

```python

def euclidian_distance(x,y):
    return np.sqrt(sum((x-y)**2))

def KNN(X,Y,query,k):
    
    m=X.shape[0]
    vals=[]
    
    for i in range(m):
        d=euclidian_distance(X[i],query)
        vals.append((d,Y[i]))
    
    vals=sorted(vals)
    vals=vals[:k]
    
    vals=np.array(vals)
    
    new_vals=np.unique(vals[:,1],return_counts=True)
    
    index=new_vals[1].argmax()
    
    pred=new_vals[0][index]
    return pred
```

```python
ans=KNN(X,Y,query,5)
print(ans)
```
