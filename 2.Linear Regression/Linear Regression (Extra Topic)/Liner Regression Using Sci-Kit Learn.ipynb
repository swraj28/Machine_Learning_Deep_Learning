```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
#Interactive plot using matplotlib
# %matplotlib notebook 
```

```python
X,y=make_regression(n_samples=1000,n_features=2,n_informative=2,noise=10,random_state=1)

print(X.shape,y.shape)
```

```python
plt.style.use("seaborn")
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from mpl_toolkits import mplot3d

fig=plt.figure()
ax=plt.axes(projection="3d")

# Creating plot
ax.scatter3D(X[:,0],X[:,1],y,color='green')
plt.title("3D scatter Plot")
plt.show()
```

### Implementing Linear Regression Using Sk-Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_ # Theta0
```

```python
model.predict(X)
```

```python
model.score(X,y)
```
