```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib notebook
```

```python
X = np.arange(20)
print(X)
```

```python
# Bias is 2 and slope is 3
"""
theta0+theta1x
"""
Theta = np.array([2,3])

noise = 3*np.random.randn(20)
```

```python
y_ideal = Theta[1]*X + Theta[0]
Y_real = Theta[1]*X + Theta[0] + noise
```

```python
plt.plot(X,y_ideal,color='orange')
plt.scatter(X,Y_real)
plt.scatter(X,noise)
plt.show()
```
