### Surface Plots | Data Visualisation

Surface Plots are used to 
- Visualise Loss Functions in Machine Learning & Deep Learning
- Visualise State or State Value Functions in Reinforcement Learning

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6,7,8])
a,b=np.meshgrid(a,b)
print(a) # a is going to be repeated b no. of times row wise.
print(b) # b is going to be repeated a no. of times column wise.
# Both a and b are of the same size and the point (x,y) represented in 2d as
# ith point of a to the ith point b.
```

```python
#Numbers in the range drom -1 to 1 having diff 0.02
a = np.arange(-1,1,0.02)
b = a 
a,b = np.meshgrid(a,b)
```

```python
print(a.shape,b.shape)
```

```python
m = a.shape[0]
n = a.shape[1]
print(m,n)

for i in range(m):
    for j in range(n):
        print("(%d,%d)"%(b[i,j],a[i,j]),end=" ")
    print()
```

```python
fig = plt.figure(figsize=(10,10))
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap='rainbow')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.title("Contour Plot")
plt.show()
```

```python
fig = plt.figure()
# axes = fig.gca(projection='2d')
plt.style.use("seaborn")
plt.contour(a,b,a**2+b**2,cmap='rainbow')
plt.title("2d Contour Plot (top view)")
plt.show()
```
